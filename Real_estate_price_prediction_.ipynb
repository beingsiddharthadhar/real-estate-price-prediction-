```python
def convertrange(x):
    temp=x.split('-')
    if len(temp)==2:
        return (float(temp[0])+float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df['total_sqft']=df['total_sqft'].apply(convertrange)
```

```python
def remove_outliers(dn):
    dn_output=pd.DataFrame()
    for key,subdn in dn.groupby('location'):
        m=np.mean(subdn.price_per_sqft)
        st=np.std(subdn.price_per_sqft)
        gen_dn=subdn[(subdn.price_per_sqft>(m-st)) & (subdn.price_per_sqft<=(m+st))]
        dn_output=pd.concat([dn_output,gen_dn],ignore_index=True)
    return dn_output
df=remove_outliers(df)
df.describe()

```

```python
def bhk_outlier_remover(dn):
    exclude_indices=np.array([])
    for location,location_dn in dn.groupby('location'):
        bhk_stats={}
        for bhk, bhk_dn in location_dn.groupby('bhk'):
            bhk_stats[bhk]={
                'mean':np.mean(bhk_dn.price_per_sqft),
                'std':np.std(bhk_dn.price_per_sqft),
                'count':bhk_dn.shape[0]
            }
        for bhk,bhk_dn in location_dn.groupby('bhk'):
            stats=bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices=np.append(exclude_indices,bhk_dn[bhk_dn.price_per_sqft<(stats['mean'])].index.values)
    return dn.drop(exclude_indices,axis='index')
```

```python
df=bhk_outlier_remover(df)
```

```python
x=df.drop(columns=['price'])
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
columntrans=make_column_transformer((OneHotEncoder(sparse=False), ['location']),remainder='passthrough')
```

```python
scaler=StandardScaler()
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(columntrans,scaler,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred_lr)
```

```python
lasso=Lasso()
```

```python
pipe=make_pipeline(columntrans,scaler,lasso)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lasso=pipe.predict(x_test)
r2_score(y_test,y_pred_lasso)
```

```python
ridge=Ridge()
```

```python
pipe=make_pipeline(columntrans,scaler,ridge)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_ridge=pipe.predict(x_test)
r2_score(y_test,y_pred_ridge)
```

```python
print("No regulariztion",r2_score(y_test,y_pred_lr))
print("Lasso",r2_score(y_test,y_pred_lasso))
print("Ridge",r2_score(y_test,y_pred_ridge))
```
